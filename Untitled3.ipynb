```python
import google.generativeai as genai
from google.colab import userdata

# To resolve 'SecretNotFoundError: Secret GEMINI_API_KEY does not exist.':
# 1. Click the '🔑' icon (Secrets) on the left sidebar of Colab.
# 2. Add a new secret named 'GEMINI_API_KEY' with your actual Gemini API key as its value.
# 3. Ensure 'Notebook access' is enabled for this secret.
api_key = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=api_key)
model = genai.GenerativeModel('gemini-2.5-flash')

print("✅ Gemini connected successfully!")

# Ensure any previous definition of analyze_data is removed
if 'analyze_data' in globals():
    del analyze_data

# Updated analyze function for Gemini
def analyze_data(df, question):
    data_summary = f"""
    Rows: {len(df)}, Columns: {list(df.columns)}
    First 5 rows:
    {df.head().to_string()}
    Statistics:
    {df.describe().to_string()}
    """
    prompt = f"You are a data analyst.\n\nData:\n{data_summary}\n\nQuestion: {question}\n\nAnswer clearly with numbers."

    response = model.generate_content(prompt)
    return response.text

print("✅ analyze_data() ready!")
```

```python
import pandas as pd

# Sample sales data
df = pd.DataFrame({
    'Month': ['Jan','Feb','Mar','Apr','May','Jun'],
    'Sales': [12000, 15000, 9000, 18000, 22000, 17000],
    'Expenses': [8000, 9000, 7500, 11000, 13000, 10000],
    'Region': ['North','South','North','East','West','South']
})

print("✅ Sample data loaded!")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def generate_chart(df, question):
    """Automatically creates the best chart for your question"""

    # Ask Gemini what chart type to use
    chart_prompt = f"""
    Given this data with columns {list(df.columns)} and question: "{question}"

    Respond with ONLY one of these options (nothing else):
    - bar_chart
    - line_chart
    - scatter_plot
    - histogram
    - pie_chart
    """

    response = model.generate_content(chart_prompt)
    chart_type = response.text.strip()

    # Draw the chart
    plt.figure(figsize=(10, 5))
    numeric_cols = df.select_dtypes(include='number').columns.tolist()

    if chart_type == "bar_chart" and len(numeric_cols) >= 1:
        df[numeric_cols[0]].plot(kind='bar', color='steelblue')
        plt.title(f"Bar Chart: {numeric_cols[0]}")

    elif chart_type == "line_chart" and len(numeric_cols) >= 1:
        df[numeric_cols].plot(kind='line', marker='o')
        plt.title("Line Chart")

    elif chart_type == "histogram" and len(numeric_cols) >= 1:
        df[numeric_cols[0]].hist(bins=15, color='coral', edgecolor='white')
        plt.title(f"Histogram: {numeric_cols[0]}")

    elif chart_type == "scatter_plot" and len(numeric_cols) >= 2:
        plt.scatter(df[numeric_cols[0]], df[numeric_cols[1]], alpha=0.6)
        plt.xlabel(numeric_cols[0]); plt.ylabel(numeric_cols[1])

    elif chart_type == "pie_chart" and len(numeric_cols) >= 1:
        df[numeric_cols[0]].plot(kind='pie', autopct='%1.1f%%')
        plt.title(f"Pie Chart: {numeric_cols[0]}")

    plt.tight_layout()
    plt.show()
    print(f"📊 Chart type chosen by AI: {chart_type}")

print("✅ generate_chart() function is ready!")
```
